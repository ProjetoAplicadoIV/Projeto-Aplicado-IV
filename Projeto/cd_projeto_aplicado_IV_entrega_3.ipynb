# **Introdução**

A poluição do ar é um problema sério que causa uma variedade de questões sociais e ambientais abrangentes. Quando se trata de saúde pública, a exposição prolongada a poluentes atmosféricos está diretamente ligada ao agravamento de várias doenças respiratórias e cardiovasculares, e pode levar à morte em casos mais graves. Além disso, o impacto no meio ambiente é significativo, pois a poluição do ar contamina corpos d'água, solo e ecossistemas naturais, prejudica a agricultura e acelera a deterioração das infraestruturas urbanas.

Identificar os tipos de substâncias poluentes presentes na atmosfera e em que quantidade é um passo essencial para lidar com esse desafio ambiental urgente. A qualidade do ar em uma determinada área pode variar consideravelmente conforme sua proximidade de fontes geradoras de poluentes, das condições climáticas e da topografia local. A partir da detecção de uma má qualidade do ar, torna-se viável investigar suas origens e, portanto, eventualmente voltar-se para o planejamento e implementação de estratégias efetivas para controlar as emissões poluentes. É fundamental conscientizar e alertar as pessoas sobre os perigos para reduzir os impactos na saúde pública.

Desde o ano de 2015, o Instituto de Energia e Meio Ambiente (IEMA) tem mantido a Plataforma da Qualidade do Ar, que busca promover a transparência em relação a esses dados ao disponibilizar informações e análises sobre o monitoramento da qualidade do ar no Brasil. Utilizando fontes acessíveis ao público, tanto relatórios estaduais anuais como fontes públicas, os dados possibilitam consultas através de índices, como o Índice de Qualidade do Ar (IQAr), bem como comparações com os padrões nacionais e diretrizes estabelecidas pela Organização Mundial da Saúde (OMS).

Neste projeto em questão, foi utilizado um conjunto de dados que adere à ideia de supervisão, contendo informações minuciosas referentes às quantidades de substâncias poluentes na atmosfera em diferentes locais e momentos específicos do dia.

A finalidade desta pesquisa é investigar e examinar tais dados, a fim de identificar padrões temporais e geográficos na presença de substâncias poluentes, levando em consideração elementos como condições climáticas e a proximidade de fontes emissoras. Por meio dessas análises, almejamos fornecer percepções que possam contribuir para o desenvolvimento e aprimoração de políticas ambientais, além de promover a conscientização acerca dos impactos da poluição do ar na saúde e no ambiente.

# **Referencial Teórico**

Os temas principais discutidos neste estudo incluem a poluição do ar - seus impactos na saúde pública e no meio ambiente - bem como a monitorização da qualidade do ar. A poluição atmosférica é caracterizada pela presença de substâncias nocivas na atmosfera - como partículas finas de material particulado ou dióxido de nitrogênio -, que estão diretamente ligadas ao agravamento de problemas respiratórios e cardiovasculares. Monitorar a qualidade do ar através de ferramentas como a disponibilizada pelo Instituto de Energia e Meio Ambiente (IEMA) é fundamental para detectar padrões de poluição ao long​ o do tempo e em diferentes regiões geográficas​, possibilitando a adoção de medidas políticas mais eficientes.

Vários estudos já examinaram os efeitos da poluição do ar e as técnicas para avaliar a sua qualidade. Além das informações da IEMA, uma pesquisa importante envolve a comparação entre os registros históricos de qualidade do ar da Agência de Proteção Ambiental (EPA) e os dados sobre casos de câncer do Instituto Nacional do Câncer dos EUA (NIH), conforme demonstrado por Jennifer Santiago no Kaggle. A pesquisa procura investigar se há uma ligação entre a qualidade do ar e o surgimento de câncer, examinando se há indicações de que a poluição do ar está associada ao aumento de casos de câncer nos Estados Unidos.

Para esta análise específica de Jennifer foram selecionadas medidas específicas focando no ozônio assim como nas partículas finas (PM2). Além disso a temperatura foi considerada pois exerce influência direta na poluição principalmente no caso do ozônio. Embora haja outros poluentes importantes como o chumbo a cobertura geográfica desses dados é menor o que tornar a análise menos robusta.

Existem opções diferentes para avaliar a qualidade do ar que incluem tanto medições feitas manualmente quanto automaticamente. As medições automáticas utilizadas por instituições como o IEMA e a EPA têm benefícios ao fornecer dados em tempo real e de forma contínua para um monitoramento mais preciso das condições atmosféricas de maneira imediata. Além disso - análises comparativas como as desenvolvidas pela EPA e NIH oferecem uma visão mais completa sobre os impactos da exposição prolongada à poluição. Contudo - é importante notar que esses sistemas automáticos enfrentam desafios relacionados aos custos elevados tanto para sua implementação quanto manutenção. Medições feitas manualmente podem ser mais fáceis de realizar; no entanto, sua precisãoe alcance temporais podem não ser comparáveis aos obtidos por outros meios automatizados mais sofisticados. Essas limitações em termos de exatidão e cobertura temporal podem diminuir a eficiência de resposta em cenários críticos.

Sazonalidade e Tendências Temporais: Identificar padrões temporais (diários, semanais, sazonais) na concentração de cada poluente usando a coluna Data_Hora.

6. Modelagem Preditiva
Previsão de Concentração de Poluentes: Usar modelos de séries temporais (ARIMA, SARIMA) ou machine learning (regressão linear) para prever a concentração de poluentes futuros com base nos dados históricos.

7. Visualização e Relatórios
Gráficos de Linhas: Visualizar a variação da concentração de poluentes ao longo do tempo para cada estação.

```python
dataframes = []

# Iterar pelos arquivos na pasta
for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        file_path = os.path.join(folder_path, filename)
        try:
            df = pd.read_csv(file_path, encoding='utf-8')
        except UnicodeDecodeError:
            df = pd.read_csv(file_path, encoding='ISO-8859-1')
        dataframes.append(df)

# Concatenar todos os DataFrames em um único DataFrame final
dataset_final = pd.concat(dataframes, ignore_index=True)
```

# **Modelo base**

Aplicação de, pelo menos, um primeiro modelo (modelo de aprendizado de máquina ou modelo estatístico) que deve ser refinado até a entrega final do projeto. Apresente o método aplicado e a análise dos resultados obtidos.

```python
dataset_final.set_index('Data_Hora', inplace=True)

df_filtered = dataset_final[(dataset_final['Estacao'] == 'Americana') & (dataset_final['Poluente'] == 'MP10')]

df_filtered = df_filtered[['Valor']].resample('h').mean()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df_filtered, label='MP10')
plt.title('Concentração de MP10 ao longo do tempo')
plt.xlabel('Data e Hora')
plt.ylabel('Valor (ug/m³)')
plt.legend()
plt.show()
```

```python
df_filtered = df_filtered.dropna()
```

```python
result = adfuller(df_filtered['Valor'])
print(f'Teste ADF: {result[1]}')
```

```python
train_size = int(len(df_filtered) * 0.8)
train, test = df_filtered[:train_size], df_filtered[train_size:]
```

```python
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(train['Valor'], order=(5, 1, 2))
model_fit = model.fit()

forecast = model_fit.forecast(steps=len(test))
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
train_scaled = scaler.fit_transform(train)
test_scaled = scaler.transform(test)

def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i + seq_length])
        y.append(data[i + seq_length])
    return np.array(X), np.array(y)

seq_length = 24
X_train, y_train = create_sequences(train_scaled, seq_length)
X_test, y_test = create_sequences(test_scaled, seq_length)

model = Sequential([
    LSTM(50, activation='relu', input_shape=(seq_length, 1)),
    Dense(1)
])

model.compile(optimizer='adam', loss='mse')
model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test))
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

mse_arima = mean_squared_error(test['Valor'], forecast)
mae_arima = mean_absolute_error(test['Valor'], forecast)
print(f'MSE (ARIMA): {mse_arima}, MAE (ARIMA): {mae_arima}')

y_pred = model.predict(X_test)
y_pred_rescaled = scaler.inverse_transform(y_pred)

mse_lstm = mean_squared_error(test[seq_length:], y_pred_rescaled)
mae_lstm = mean_absolute_error(test[seq_length:], y_pred_rescaled)
print(f'MSE (LSTM): {mse_lstm}, MAE (LSTM): {mae_lstm}')
```

```python
plt.figure(figsize=(10, 5))
plt.plot(test.index, test['Valor'], label='Real')
plt.plot(test.index, forecast, label='Previsão (ARIMA)')
plt.title('Previsão de Concentração de MP10 - ARIMA')
plt.legend()
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(test.index[seq_length:], test['Valor'][seq_length:], label='Real')
plt.plot(test.index[seq_length:], y_pred_rescaled, label='Previsão (LSTM)')
plt.title('Previsão de Concentração de MP10 - LSTM')
plt.legend()
plt.show()
```
