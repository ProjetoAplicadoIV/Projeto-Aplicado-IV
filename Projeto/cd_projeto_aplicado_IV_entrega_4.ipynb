# **Introdução**

A poluição do ar é um problema sério que causa uma variedade de questões sociais e ambientais abrangentes. Quando se trata de saúde pública, a exposição prolongada a poluentes atmosféricos está diretamente ligada ao agravamento de várias doenças respiratórias e cardiovasculares, e pode levar à morte em casos mais graves. Além disso, o impacto no meio ambiente é significativo, pois a poluição do ar contamina corpos d'água, solo e ecossistemas naturais, prejudica a agricultura e acelera a deterioração das infraestruturas urbanas.

Identificar os tipos de substâncias poluentes presentes na atmosfera e em que quantidade é um passo essencial para lidar com esse desafio ambiental urgente. A qualidade do ar em uma determinada área pode variar consideravelmente conforme sua proximidade de fontes geradoras de poluentes, das condições climáticas e da topografia local. A partir da detecção de uma má qualidade do ar, torna-se viável investigar suas origens e, portanto, eventualmente voltar-se para o planejamento e implementação de estratégias efetivas para controlar as emissões poluentes. É fundamental conscientizar e alertar as pessoas sobre os perigos para reduzir os impactos na saúde pública.

Desde o ano de 2015, o Instituto de Energia e Meio Ambiente (IEMA) tem mantido a Plataforma da Qualidade do Ar, que busca promover a transparência em relação a esses dados ao disponibilizar informações e análises sobre o monitoramento da qualidade do ar no Brasil. Utilizando fontes acessíveis ao público, tanto relatórios estaduais anuais como fontes públicas, os dados possibilitam consultas através de índices, como o Índice de Qualidade do Ar (IQAr), bem como comparações com os padrões nacionais e diretrizes estabelecidas pela Organização Mundial da Saúde (OMS).

Neste projeto em questão, foi utilizado um conjunto de dados que adere à ideia de supervisão, contendo informações minuciosas referentes às quantidades de substâncias poluentes na atmosfera em diferentes locais e momentos específicos do dia.

A finalidade desta pesquisa é investigar e examinar tais dados, a fim de identificar padrões temporais e geográficos na presença de substâncias poluentes, levando em consideração elementos como condições climáticas e a proximidade de fontes emissoras. Por meio dessas análises, almejamos fornecer percepções que possam contribuir para o desenvolvimento e aprimoração de políticas ambientais, além de promover a conscientização acerca dos impactos da poluição do ar na saúde e no ambiente.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.layers import Dropout
from tensorflow.keras.optimizers import Adam

# Normalizar os dados
scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df_filtered)

# Criar sequências para treinamento
def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i + seq_length])
        y.append(data[i + seq_length])
    return np.array(X), np.array(y)

seq_length = 72  # Exemplo: usar dados das últimas 72 horas para prever a próxima
X, y = create_sequences(df_scaled, seq_length)

# Dividir em treino e teste
train_size = int(0.8 * len(X))
X_train, y_train = X[:train_size], y[:train_size]
X_test, y_test = X[train_size:], y[train_size:]

# Definir o modelo LSTM
model = Sequential()
model.add(LSTM(100, return_sequences=True, input_shape=(seq_length, 1)))
model.add(Dropout(0.2))  # Adicionando Dropout
model.add(LSTM(50, return_sequences=False))
model.add(Dropout(0.2))  # Adicionando Dropout
model.add(Dense(1))
model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')

# Callback para parar o treinamento cedo se não houver melhora
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Treinar o modelo com validação
history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test), callbacks=[early_stopping])

# Verifique as previsões futuras e faça a análise conforme descrito acima.

# Fazer previsões futuras
future_predictions = []
last_sequence = df_scaled[-seq_length:]  # Última sequência do conjunto de dados

future_steps = 24 * 30 * 3  # Prever os próximas 3 meses

for _ in range(future_steps):
    input_sequence = last_sequence.reshape(1, seq_length, 1)
    predicted_value = model.predict(input_sequence, verbose=0)
    future_predictions.append(predicted_value[0, 0])
    last_sequence = np.append(last_sequence[1:], predicted_value, axis=0)

# Desnormalizar as previsões
future_predictions_rescaled = scaler.inverse_transform(np.array(future_predictions).reshape(-1, 1))

# Criar um índice de datas para previsões futuras
future_dates = pd.date_range(start=df_filtered.index[-1], periods=future_steps + 1, freq='H')[1:]
```

O modelo LSTM (Long Short-Term Memory) desenvolvido para esta tarefa é uma variação avançada de redes neurais recorrentes, projetada para lidar com desafios em séries temporais que apresentam dependências de longo prazo. A estrutura de uma LSTM é composta de células de memória e mecanismos de portas que permitem ao modelo reter ou descartar informações conforme necessário. Essa habilidade de armazenar e manipular contextos temporais complexos torna as LSTMs particularmente úteis em tarefas onde os padrões temporais e sequenciais precisam ser capturados, como previsão de séries temporais financeiras, meteorológicas ou, neste caso, de dados sequenciais de grande volume.

A arquitetura específica deste projeto foi projetada para equilibrar a capacidade de aprendizado com a mitigação de overfitting. A camada inicial de 100 unidades LSTM foi configurada para retornar sequências completas, permitindo que a segunda camada LSTM receba uma sequência estruturada de informações. Essa abordagem ajuda a rede a captar padrões mais detalhados e relacionamentos temporais mais profundos nos dados. O uso do return_sequences=True na primeira camada é uma estratégia importante quando se utiliza múltiplas camadas LSTM, pois permite que cada camada subsequente processe sequências completas em vez de apenas o último estado de saída.

Para melhorar a robustez do modelo, foram adicionadas camadas de Dropout com uma taxa de 20% após as camadas LSTM. O Dropout é uma técnica de regularização que desativa aleatoriamente uma fração de neurônios durante o treinamento, evitando que o modelo se ajuste excessivamente aos dados de treinamento. Esta prática ajuda a rede a generalizar melhor para novos dados, reduzindo o risco de overfitting. Na segunda camada LSTM, que contém 50 unidades e é configurada para não retornar sequências, o Dropout também contribui para uma redução eficaz da complexidade do modelo e aprimora sua capacidade de generalização.

A camada de saída do modelo é uma camada Dense com um único nó, que é responsável por gerar a previsão final para a próxima etapa da série temporal. A escolha de apenas um nó na saída é apropriada para modelos de previsão de séries temporais univariadas, onde a tarefa é prever um único valor por vez.

O otimizador Adam foi utilizado por sua eficiência e capacidade de ajuste adaptativo de taxa de aprendizado. O Adam combina as vantagens dos otimizadores RMSprop e SGD com momentum, resultando em um algoritmo que ajusta a taxa de aprendizado com base na primeira e segunda momenta do gradiente. Definir a taxa de aprendizado em 0.001 é uma escolha comum, pois oferece um bom equilíbrio entre a velocidade de convergência e a estabilidade do treinamento. Esta configuração é particularmente útil em redes LSTM, onde taxas de aprendizado muito altas podem levar a oscilações e divergências, enquanto taxas muito baixas podem resultar em um processo de treinamento lento.

Para garantir que o modelo não seja treinado além do necessário, foi utilizado o callback de EarlyStopping, monitorando a métrica de val_loss. Este callback interrompe o treinamento se não houver melhora na perda de validação por 5 épocas consecutivas, restaurando automaticamente os pesos do modelo para o estado em que a perda foi mínima. O uso do restore_best_weights=True é crucial, pois assegura que o modelo resultante esteja na melhor versão encontrada durante o treinamento, maximizando seu desempenho em dados de teste e futuros.

O pré-processamento dos dados envolveu a normalização usando MinMaxScaler, que mapeia os valores para um intervalo entre 0 e 1. Este procedimento é essencial, pois as redes LSTM são sensíveis à escala dos dados, e trabalhar com valores normalizados melhora a eficiência do treinamento e a estabilidade numérica do modelo. A função de criação de sequências foi configurada para gerar janelas deslizantes de 72 horas de dados, que servem como entrada para a rede, e a saída associada é o valor imediatamente após essa sequência. Essa abordagem permite que o modelo considere padrões mais amplos e obtenha uma visão detalhada das tendências e flutuações dentro de um período de três dias.

A divisão dos dados em conjuntos de treinamento e teste foi feita em uma proporção de 80% para treinamento e 20% para teste, garantindo que a rede seja avaliada em dados que ela não viu durante o processo de aprendizado. Isso é importante para validar a capacidade de generalização do modelo e assegurar que ele não esteja simplesmente "decorando" os dados de treinamento, mas sim aprendendo padrões relevantes.

As previsões futuras foram geradas de forma iterativa, começando com a última sequência do conjunto de dados como entrada. O modelo foi utilizado para prever o próximo valor, que em seguida foi adicionado à sequência, formando a entrada para a próxima previsão. Este processo foi repetido até que um período de três meses (24 horas x 30 dias x 3) fosse preenchido. A abordagem iterativa é comum em séries temporais e permite ao modelo criar previsões de longo prazo que levam em conta suas próprias saídas anteriores.

Após a geração das previsões, os valores previstos foram desnormalizados aplicando a transformação inversa do MinMaxScaler, para trazer os dados de volta à escala original. Essa etapa é necessária para interpretar as previsões de maneira prática e fazer análises comparativas com os dados históricos reais. O resultado final foi então plotado para visualização e análise, fornecendo uma visão clara de como o modelo antecipa os movimentos da série temporal ao longo dos próximos meses.
